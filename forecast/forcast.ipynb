# Load Data

```python
import pandas as pd
df = pd.read_excel("./data/completion.xlsx", engine = "openpyxl")
```

# Strategy 1. Mean of Week
주 주기성을 띄는 식수인원의 특성을 이용해서, 요일별 평균 식수인원을 이용해 예측하고 그 결과를 비교해본다.

```python
df_week_mean = df.groupby(['Week','Hour'])['HeadCount'].agg(**{'mean_count':'mean',
                                       'min_count':'min',
                                       'max_count':'max'}).reset_index().sort_values(by=['mean_count'], 
                                                                                     axis=0,
                                                                                     ascending=False)

_is_week = df_week_mean["Week"] == "월"
_is_hour = df_week_mean["Hour"] == 11
#df_week_mean[_is_week & _is_hour]["mean_count"]
df_week_mean
```

```python
df_pred = df.loc[:,["Date","Hour","Week","HeadCount"]]
for i in df_pred.index:
    _is_week = (df_week_mean["Week"] == df_pred["Week"][i])
    _is_hour = (df_week_mean["Hour"] == df_pred["Hour"][i])
    df_pred["HeadCount"][i] = df_week_mean[_is_week & _is_hour].copy()["mean_count"]
df_pred.head()
```

```python
df.head()
```

```python
total_df = pd.merge(df,df_pred,on=["Date", "Hour"])
total_df.drop(["Week_y"], axis=1)
total_df.to_csv("./data/pred_1.csv", mode="w") # save result
```

```python
import matplotlib.pyplot as plt
plt.plot(df_pred["Date"],df_pred["HeadCount"])
plt.plot(df_pred["Date"],df["HeadCount"])
plt.show()
```

```python
# Check Error
import numpy as np
actual = df["HeadCount"].to_numpy()
fake = df_pred["HeadCount"].to_numpy()
diff = [ (actual[n] - fake[n]) for n in range(len(actual))]
print("Abs Diff Mean",np.nanmean(diff))
print("Abs Diff Mean",np.nanmean(list(map(abs,diff))))
```

# Strategy 2. TSL Result
TSL 결과로 나온 값은 다음과 같다.

```python
hour_meal = {"7":"breakfast", "11":"lunch", "17":"dinner"}
seasonal = {
    "월": {"breakfast": 21.34717608, "lunch":63.04229665,"dinner":43.46543268},
    "화": {"breakfast": 23.41694352, "lunch":34.803094,"dinner":41.64151242},
    "수": {"breakfast": 26.4833887, "lunch":10.04561891,"dinner":44.07340611},
    "목": {"breakfast": 8.33594368, "lunch":21.89928131,"dinner":9.975478564},
    "금": {"breakfast": -3.055212783, "lunch":58.89587995,"dinner":-35.55513368},
    "토": {"breakfast": -34.46677741, "lunch":-99.12049404,"dinner":-69.17908559},
    "일": {"breakfast": -42.06146179, "lunch":-89.56567677,"dinner":-34.4216105},
}
trend = {"breakfast": 86.29049212, "lunch":225.8122312,"dinner":187.242236}
```

```python
df_pred = df.loc[:,["Date","Hour","Week","HeadCount"]]
for i in df_pred.index:
    df_pred["HeadCount"][i] = seasonal[df_pred["Week"][i]][hour_meal[str(df_pred["Hour"][i])]] + trend[hour_meal[str(df_pred["Hour"][i])]]
df_pred.head()
```

```python
total_df = pd.merge(df,df_pred,on=["Date", "Hour"])
total_df.drop(["Week_y"], axis=1)
total_df.to_csv("./data/pred_2.csv", mode="w")
import matplotlib.pyplot as plt
plt.plot(df_pred["Date"],df_pred["HeadCount"])
plt.plot(df_pred["Date"],df["HeadCount"])
```

```python
# Check Error
import numpy as np
actual = df["HeadCount"].to_numpy()
fake = df_pred["HeadCount"].to_numpy()
diff = [ (actual[n] - fake[n]) for n in range(len(actual))]
print("Abs Diff Mean",np.nanmean(diff))
print("Abs Diff Mean",np.nanmean(list(map(abs,diff))))
```
